```python
csv_file=open("OLX_mobile_data.csv","w",newline="")
write=csv.writer(csv_file)
write.writerow(["Mobile discription","Price"])
for data in soup.find_all("div",class_="IKo3_"):
    try:
        name=data.find("span",class_="_2tW1I")
        name=name.text
        
    except:
        print("-name is not present-\n")
    else:
        print("this is the name: \n",name)
    try:
        price=data.find("span",class_="_89yzn")
        price=price.text
    except:
        print("-price is not present-\n")
    else:
        print("this is the price: \n",price)
    row=[name,price]
    write.writerow(row)
csv_file.close()    

    
```
